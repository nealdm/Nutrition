```python
df.rename(columns = {c:c[6:-5] for c in [i for i in df.columns if i[-4:] == "(IU)"]},inplace = True)
```

```python
# g/IU
VitaminA_conv = 0.6/1000000
VitaminD_conv = 0.025/1000000
```

```python
# df.rename(columns = {c:c[6:-5] for c in [i for i in df.columns if i[-4:] == "(IU)"]},inplace = True)
df.rename(columns = {c:c[6:-4] for c in [i for i in df.columns if i[-3:] == "(g)"]},inplace = True)
```

```python
dfmean = df.groupby("category").mean()
```

```python
list(dfmean.columns)
```

```python
sub_df = dfmean.sort_values("Polyunsaturated fatty acids",ascending= False)
sub_df = sub_df[:10].copy()

index = sub_df.index.values

dfplots = pd.DataFrame({'Polyunsaturated fatty acids': sub_df['Polyunsaturated fatty acids'].values,
                        'Monounsaturated fatty acids':sub_df['Monounsaturated fatty acids'].values,
                        'Fatty acids, total trans': sub_df['Fatty acids, total trans'].values,
                   'Saturated fatty acids': sub_df['Saturated fatty acids'].values}, index=index)
ax = dfplots.plot.bar(rot=45)
```
